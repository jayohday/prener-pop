```python
import pandas as pd
```

```python
df_race = pd.read_csv('../STL_DEMOGRAPHY_NhoodPop/data/clean/STL_RaceByNhood.csv')
df_pop = pd.read_csv('../STL_DEMOGRAPHY_NhoodPop/data/clean/STL_PopByNhood.csv')
```

```python
df_pop.info()
```

```python
df_race['50_10_white_change'] = df_race['white10'] - df_race['white50']
```

```python
df_race['50_10_black_change'] = df_race['black10'] - df_race['black50']
```

```python
df_race.set_index('NHD_NAME')
df_pop.set_index('NHD_NAME')
```

```python
df_join = pd.concat([df_race, df_pop], axis=1)
```

```python
df_join.info()
```

```python
new_df_join = df_join.loc[:,~df_join.T.duplicated(keep='first')]
```

```python
### Black population
df['40_50_black'] = df['black50'] - df['nonwhite40']
df['50_60_black'] = df['black60'] - df['black50']
df['60_70_black'] = df['black70'] - df['black60']
df['70_80_black'] = df['black80'] - df['black70']
df['80_90_black'] = df['black90'] - df['black80']
df['90_00_black'] = df['black00'] - df['black90']
df['00_10_black'] = df['black10'] - df['black00']
df['10_11_black'] = df['black11'] - df['black10']
df['11_12_black'] = df['black12'] - df['black11']
df['12_13_black'] = df['black13'] - df['black12']
df['13_14_black'] = df['black14'] - df['black13']
df['14_15_black'] = df['black15'] - df['black14']
df['15_16_black'] = df['black16'] - df['black15']
df['16_17_black'] = df['black17'] - df['black16']
df['50_00_black'] = df['black00'] - df['black50']
df['00_17_black'] = df['black17'] - df['black00']
df['40_17_black'] = df['black17'] - df['nonwhite40']
```

```python
### Population figures
df['40_50_pop'] = df['pop50'] - df['pop40']
df['50_60_pop'] = df['pop60'] - df['pop50']
df['60_70_pop'] = df['pop70'] - df['pop60']
df['70_80_pop'] = df['pop80'] - df['pop70']
df['80_90_pop'] = df['pop90'] - df['pop80']
df['90_00_pop'] = df['pop00'] - df['pop90']
df['00_10_pop'] = df['pop10'] - df['pop00']
df['10_11_pop'] = df['pop11'] - df['pop10']
df['11_12_pop'] = df['pop12'] - df['pop11']
df['12_13_pop'] = df['pop13'] - df['pop12']
df['13_14_pop'] = df['pop14'] - df['pop13']
df['14_15_pop'] = df['pop15'] - df['pop14']
df['15_16_pop'] = df['pop16'] - df['pop15']
df['16_17_pop'] = df['pop17'] - df['pop16']
df['50_00_pop'] = df['pop00'] - df['pop50']
df['00_17_pop'] = df['pop17'] - df['pop00']
df['40_17_pop'] = df['pop17'] - df['pop40']
df['50_10_pop'] = df['pop10'] - df['pop50']
```
